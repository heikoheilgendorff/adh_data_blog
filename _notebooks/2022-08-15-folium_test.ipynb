```python
data_path = '../../spyder/data/'
df_0 = pd.read_csv('%szaf_general_2020_csv.zip'% data_path)
df_1 = pd.read_csv('%szaf_men_2020_csv.zip'% data_path)
df_2 = pd.read_csv('%szaf_women_2020_csv.zip'% data_path)

df = pd.merge(df_0,df_1,how='left',on=['longitude','latitude'])
df = pd.merge(df,df_2,how='left',on=['longitude','latitude'])
#%%
# CRS = EPSG:4326
def create_map(df):
    gdf_map = gpd.GeoDataFrame(df,geometry = gpd.points_from_xy(df.longitude, df.latitude, crs="EPSG:4326"))
    return gdf_map#.iloc[:,[2,3,4,5]]

gdf_0 = create_map(df)
#gdf_0.plot()
df_0_h = df_0.head(100)
df_1_h = df_1.head(100)
df_2_h = df_2.head(100)
df_h = df.head(100)
#%%
gdf = gpd.read_file('%sgeo/zaf_adm_sadb_ocha_20201109_SHP/zaf_admbnda_adm4_sadb_ocha_20201109.shp'% data_path)
#gdf = gpd.read_file('./data/geo/zaf_adm_sadb_ocha_20201109_SHP/zaf_admbndp_admALL_sadb_ocha_itos_20201109.shp')
#gdf.plot()
prov = gdf['ADM1_EN'].drop_duplicates().to_list()
gdf_h = gdf.head(2)

gdf_kzn = gdf[gdf.ADM1_EN=='KwaZulu-Natal']
#gdf_kzn.plot()
#%%
def get_prov(gdf,gdf_data):
    gdf = gdf[gdf.ADM3_EN=='eThekwini']
    gdf_prov_1 = gpd.sjoin(gdf,gdf_data)
    gdf_prov = gpd.sjoin(gdf_data,gdf)
    return gdf_prov_1,gdf_prov

gdf_prov_1,gdf_prov = get_prov(gdf_kzn,gdf_0)


#ax= gdf_kzn.boundary.plot()
#gdf_prov.plot(ax=ax,column='zaf_general_2020')


#gdf_prov.plot(column='zaf_general_2020')


#%%



```
